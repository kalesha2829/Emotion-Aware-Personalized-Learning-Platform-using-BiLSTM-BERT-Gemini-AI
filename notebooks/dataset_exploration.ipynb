# Dataset Collection & Exploration

**AI Learning Assistant** — This notebook loads the [GoEmotions](https://www.kaggle.com/datasets/dehbmultani/go-emotions-dataset) dataset, maps its 27 fine-grained emotion labels to five student learning emotions, and performs exploratory data analysis.

**Target emotions:** Bored, Confident, Confused, Curious, Frustrated

> **Note:** No text preprocessing, tokenization, or model training is performed in this module.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd

from utils.dataset_loader import (
    apply_target_mapping,
    clean_raw_dataset,
    get_dataset_metadata,
    load_raw_dataset,
    save_cleaned_dataset,
)
from utils.dataset_explorer import (
    check_duplicates,
    check_missing_values,
    display_head,
    display_info,
    generate_observations,
    show_label_distribution,
    show_sample_count,
)
from utils.emotion_mapping import print_mapping_summary, MAPPING_DOCUMENTATION
from utils.visualizations import create_all_visualizations
```

## 1. Dataset Selection & Label Mapping

We use **GoEmotions** (Google Research), a Kaggle-hosted corpus of ~58k Reddit comments labeled with 27 fine-grained emotions. Three target classes map directly (`confusion`, `curiosity`, `annoyance`/`anger`). `Bored` uses proxy labels because boredom was removed from the GoEmotions taxonomy during annotation design.

```python
metadata = get_dataset_metadata()
print(f"Dataset : {metadata['name']}")
print(f"Kaggle  : {metadata['kaggle_url']}")
print(f"HF ID   : {metadata['huggingface_id']}")
print_mapping_summary()
pd.DataFrame([
    {"Target": k, "Source Labels": ", ".join(v["source_labels"]), "Rationale": v["rationale"]}
    for k, v in MAPPING_DOCUMENTATION.items()
])
```

## 2. Load Dataset

```python
raw_df = load_raw_dataset()
mapped_df = apply_target_mapping(raw_df)
df = clean_raw_dataset(mapped_df)
print(f"Loaded {len(df):,} cleaned samples.")
```

## 3. Exploratory Data Analysis

```python
display_head(df)
display_info(df)
duplicate_stats = check_duplicates(df)
check_missing_values(df)
show_sample_count(df)
distribution = show_label_distribution(df)
```
